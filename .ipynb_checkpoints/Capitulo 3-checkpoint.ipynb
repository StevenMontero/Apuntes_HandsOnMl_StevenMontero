### Measuring Accuracy Using Cross- Validation 

#### Implementaci√≥n de Cross validation 

De vez en cuando, necesitar√° m√°s control sobre el proceso de validaci√≥n cruzada que el que ofrece Scikit-Learn. En estos casos, puede implementar la validaci√≥n cruzada usted mismo. El siguiente c√≥digo hace aproximadamente lo mismo que la funci√≥n cross_val_score () de Scikit-Learn e imprime el mismo resultado:

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.base import clone 
# Clone hace una copia profunda del modelo en un estimador sin copiar los datos adjuntos. 
#Produce un nuevo estimador con los mismos par√°metros que no se ha ajustado a ning√∫n dato.


skfolds = StratifiedKFold(n_splits=3,random_state=42)

# train_index es un array con los indices de valores a usar para el entranamiento.
for train_index, test_index in skfolds.split(X_train,y_train_5):
    clone_clf = clone(sgd_clf)
    X_train_folds = X_train[train_index]
    y_train_folds = y_train_5[train_index]
    X_test_fold = X_train[test_index]
    y_test_fold = y_train_5[test_index]
    
    clone_clf.fit(X_train_folds, y_train_folds)
    y_pred = clone_clf.predict(X_test_fold)
    n_correct = sum(y_pred == y_test_fold)
    print(n_correct / len(y_pred))
```

#### Exactitud / accuracy 
La exactitud se refiere a qu√© tan cerca est√°n del valor real las mediciones de un sistema de medici√≥n.


#### Precisi√≥n /  precision
La precisi√≥n se refiere a qu√© tan cerca est√°n las mediciones entre ellas. 

 

```python
# Con la clase de sklearn
from sklearn.model_selection import cross_val_score
cross_val_score(sgd_clf, X_train, y_train_5, cv=3, scoring="accuracy")
```

¬°Guau! ¬øM√°s del 93% de precisi√≥n (proporci√≥n de predicciones correctas) en todos los pliegues de validaci√≥n cruzada? Esto se ve incre√≠ble, ¬øno? Bueno, antes de emocionarse demasiado, veamos un clasificador muy tonto que solo clasifica cada imagen en la clase "no 5":

```python
from sklearn.base import BaseEstimator

class Nerver5Classifier(BaseEstimator):
    def fit(self, X, y=None):
        return self
    def predict(self,X):
        return np.zeros((len(X),1),dtype=bool)
```

¬øPuedes adivinar la precisi√≥n de este modelo? Vamos a averiguar:

```python
never_5_clf = Nerver5Classifier()
cross_val_score(never_5_clf,X_train,y_train_5,cv=3,scoring='accuracy')
```

As√≠ es, tiene una precisi√≥n de m√°s del 90%. Esto se debe simplemente a que solo alrededor del 10% de las im√°genes son 5, por lo que si siempre adivinas que una imagen no es un 5, acertar√°s aproximadamente el 90% de las veces.

**Esto demuestra por qu√© la *accuracy* generalmente no es la medida de rendimiento preferida para los clasificadores, especialmente cuando se trata de conjuntos de datos sesgados(skewed datasets) es decir, cuando algunas clases son mucho m√°s frecuentes que otras.**

###  Confusion Matrix

### Metricas para evaluar el rendimiento de los modelos de clasificacion 

#### Exactitud / accuracy
Es el n√∫mero de predicciones correctas realizadas por el modelo por el n√∫mero total de registros. La mejor precisi√≥n es el 100%, lo que indica que todas las predicciones son correctas.
Toma en cuenta que la exactitud no es una medida v√°lida del rendimiento del modelo cuando se tiene un conjunto de datos desequilibrado. 



#### Precisi√≥n / precision
Con precisi√≥n, estamos evaluando nuestros datos por su desempe√±o de predicciones ‚Äúpositivas‚Äù. En otras palabras es el n√∫mero de elementos identificados correctamente como positivo de un total de elementos identificados como positivos(que el modelo predijo como positivos pero que lagunos no lo son).


#### Sensibilidad / Recall
La sensibilidad/Recall se calcula como el n√∫mero de predicciones positivas correctas dividido por el n√∫mero total de positivos. En otras palabras es el n√∫mero de elementos identificados correctamente como positivos del total de positivos verdaderos.
 

#### Especificidad
La especificidad, tasa negativa verdadera, se calcula como el n√∫mero de predicciones negativas correctas dividido por el n√∫mero total de negativos verdaderos. Lo contrario al Recall. En otras palabras es el n√∫mero de elementos identificados correctamente como negativos del total de negativos verdaderos.

 

#### Puntaje de F1
Realmente no queremos calcular la precisi√≥n y la sensibilidad cada vez que creamos un modelo para resolver un problema de clasificaci√≥n, as√≠ que es mejor si podemos obtener una puntuaci√≥n √∫nica que represente ambas variables.

El puntaje F1 es el promedio ponderado de precisi√≥n y sensibilidad. Por lo tanto, esta puntuaci√≥n tiene en cuenta tanto los falsos positivos como los falsos negativos.

Entonces, si un n√∫mero es realmente peque√±o entre la precisi√≥n y el recuerdo, el puntaje F1 levanta una bandera y est√° m√°s cerca del n√∫mero m√°s peque√±o que el m√°s grande, dando al modelo un puntaje apropiado en lugar de solo una media aritm√©tica.

La media arm√≥nica es una especie de promedio cuando xey son iguales. Pero cuando xey son diferentes, entonces est√° m√°s cerca del n√∫mero m√°s peque√±o en comparaci√≥n con el n√∫mero m√°s grande.



Una forma mucho mejor de evaluar el desempe√±o de un clasificador es mirar la matriz de confusi√≥n. Para calcular la matriz de confusi√≥n, primero debe tener un conjunto de predicciones para poder compararlas con los objetivos reales.Puede utilizar el **cross_val_predict().

Al igual que la funci√≥n cross_val_score (), cross_val_predict () realiza una validaci√≥n cruzada de K veces, pero en lugar de devolver los puntajes de evaluaci√≥n, devuelve las predicciones realizadas en cada pliegue de prueba. Esto significa que obtienes una predicci√≥n limpia para cada instancia en el conjunto de entrenamiento ("limpio" significa que la predicci√≥n la realiza un modelo que nunca vio los datos durante el entrenamiento).

```python
from sklearn.model_selection import cross_val_predict
y_train_pred = cross_val_predict(sgd_clf,X_train,y_train_5,cv=3)
```



Ya con las predicciones listas, se puede obtenr la matris de confusion usando **confusion_matrix()** solo ocupamos pasarle la clase o valores objetivos (y_train_5) y las prediciones que obtuvimos.

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_train_5,y_train_pred)
```

Cada fila en una matriz de confusi√≥n representa una clase actual, mientras que cada columna representa una clase predicha(los valores de la prediccion). La primera fila de esta matriz considera im√°genes que no son 5 (la clase negativa): 53.057 de ellas se clasificaron correctamente como no 5 (se denominan verdaderos negativos), mientras que las 1.522 restantes se clasificaron err√≥neamente como 5 (falsos positivos). La segunda fila considera las im√°genes de 5 (la clase positiva): 1.325 se clasificaron err√≥neamente como no 5 (falsos negativos), mientras que las 4096 restantes se clasificaron correctamente como 5 (verdaderos positivos). Un clasificador perfecto solo tendr√≠a verdaderos positivos y verdaderos negativos, por lo que su matriz de confusi√≥n tendr√≠a valores distintos de cero solo en su diagonal principal /*Nota: el orden de como esten organizadas las filas y columnas puede variar.*/ (de arriba a la izquierda a abajo a la derecha):



```python
y_train_perfect_predictions = y_train_5 # simula un caso en donde las predicciones son perfectas
confusion_matrix(y_train_5,y_train_perfect_predictions)
```

La matriz de confusi√≥n le brinda mucha informaci√≥n, pero a veces puede preferir una m√©trica m√°s concisa. Uno interesante de observar es la accuracy de las predicciones positivas; esto se llama **precision** del clasificador.
Esta dada por la formula: 

# $ precision = \frac{TP}{TP+FP}$

TP is the number of true positives, and FP is the number of false positives.


La precisi√≥n se usa t√≠picamente junto con otra m√©trica llamada **Recall**, tambi√©n llamada **Sensitivity** o **true positive rate** (TPR): esta es la proporci√≥n de instancias positivas que el clasificador detecta correctamente

# $ recall = \frac{TP}{TP+FN}$

FN is, of course, the number of false negatives.

### Sklearn nos da muchas funciones para calcular las metricas de los clasificadores, entre ellas precision y recall

A menudo es conveniente combinar precision y recall en una √∫nica m√©trica llamada puntuaci√≥n F1 score, en particular si necesita una forma sencilla de comparar dos clasificadores. La puntuaci√≥n F1 es la media arm√≥nica de precision y recall. Mientras que la media regular trata todos los valores por igual, la media arm√≥nica da mucho m√°s peso a los valores bajos. Como resultado, el clasificador solo obtendr√° una puntuaci√≥n F1 alta si tanto el recall como la precision son altas.


Para calcular el F1 Score, simplemente podemos usar la funcion f1_score() de Sklearn 

```python
from sklearn.metrics import f1_score
f1_score(y_train_5,y_train_pred)
```

La puntuaci√≥n F1 favorece a los clasificadores que tienen una precisi√≥n y una memoria similares. Esto no siempre es lo que desea: en algunos contextos, lo que m√°s le importa es la precisi√≥n, y en otros contextos, realmente le importa el recuerdo. Por ejemplo, si entren√≥ a un clasificador para detectar videos que son seguros para los ni√±os, probablemente prefiera un clasificador que rechace muchos videos buenos (poca recall) pero que solo conserve los seguros (alta precisi√≥n), en lugar de un clasificador que tenga mucho mayor recuperaci√≥n, pero permite que aparezcan algunos videos realmente malos en su producto (en tales casos, es posible que incluso desee agregar una canalizaci√≥n humana para verificar la selecci√≥n de videos del clasificador). Por otro lado, suponga que entrena a un clasificador para detectar ladrones en las im√°genes de vigilancia: probablemente est√© bien si su clasificador tiene solo un 30% de precisi√≥n siempre que tenga un 99% de recall (seguro, los guardias de seguridad recibir√°n algunas alertas falsas, pero casi todos los ladrones ser√°n atrapados).


En otras palabras: 

Es mejor mostrar pocos videos pero que sean 100% seguros para ni√±os (alta presicion) y dejar algunos videos buenos por fuera(clasificarlos como malos,bajo recall).
 
Es mejor perdir perdon cuando la alerta de ladron sea falsa (baja presicion) mientras se puedan atrapar a todos los ladrones (alto reacall).


**Desafortunadamente, no puede tener las dos cosas: aumentar la precisi√≥n reduce el recall y viceversa. Esto se denomina precision/ recall trade-off.**


 ### Precision/ recall trade-off.
 
 Para comprender este compromiso, veamos c√≥mo el clasificador SGDC toma sus decisiones de clasificaci√≥n. Para cada instancia, calcula una puntuaci√≥n basada en una funci√≥n de decisi√≥n. Si esa puntuaci√≥n es mayor que un umbral, asigna la instancia a la clase positiva; de lo contrario, lo asigna a la clase negativa. La Figura 3-3 muestra algunos d√≠gitos colocados desde la puntuaci√≥n m√°s baja a la izquierda hasta la puntuaci√≥n m√°s alta a la derecha. Suponga que el umbral de decisi√≥n se coloca en la flecha central (entre los dos 5): encontrar√° 4 verdaderos positivos (5 reales) a la derecha de ese umbral y 1 falso positivo (en realidad, un 6). Por tanto, con ese umbral, la precisi√≥n es del 80% (4 de 5). Pero de 6 de los 5-reales, el clasificador solo detecta 4, por lo que la recuperaci√≥n es del 67% (4 de 6). Si aumenta el umbral (mu√©valo a la flecha de la derecha), el falso positivo (el 6) se convierte en un verdadero negativo, lo que aumenta la precisi√≥n (hasta el 100% en este caso), pero un verdadero positivo se convierte en un falso negativo. , disminuyendo el recall al 50%. Por el contrario, reducir el umbral aumenta la recall y reduce la precisi√≥n.
 
 
 
 
 Scikit-Learn no le permite establecer el umbral directamente, pero le da acceso a las puntuaciones de decisi√≥n que utiliza para hacer predicciones. En lugar de llamar al m√©todo predict () del clasificador, puede llamar a su m√©todo decision_function (), que devuelve una puntuaci√≥n para cada instancia, y luego usar cualquier umbral que desee para hacer predicciones basadas en esas puntuaciones:

```python
y_scores = sgd_clf.decision_function([some_digit])
y_scores
```

```python
threshold = 0
y_some_digit_pred = (y_scores > threshold)
y_some_digit_pred
```

El SGDClassifier usa un umbral igual a 0, por lo que el c√≥digo anterior devuelve el mismo resultado que el m√©todo predict () (es decir, True).Pero que pasaria si elevamos el umbral:

```python
threshold = 8000
y_some_digit_pred = (y_scores > threshold)
y_some_digit_pred
```

Esto confirma que cuando se eleva el threshold el recall decrese.

¬øC√≥mo decide qu√© umbral utilizar? Primero, use la funci√≥n cross_val_predict () para obtener las puntuaciones de todas las instancias en el conjunto de entrenamiento, pero esta vez especifique que desea devolver puntuaciones de decisi√≥n en lugar de predicciones (method = 'decision_function):

```python
y_scores  = cross_val_predict(sgd_clf,X_train,y_train_5,cv=3,method='decision_function')
```

```python
y_scores
```

Con estos puntajes, use la funci√≥n precision_recall_curve () para calcular la precisi√≥n y el recall de todos los umbrales posibles

```python
from sklearn.metrics import precision_recall_curve

precisions, recalls, thresholds = precision_recall_curve(y_train_5,y_scores)
```

Para observar estos resultados de una menera mas intuitiva usemos math plot lib para graficarlo

```python
def plot_precision_recall_vs_threshold(precisions,recalls,thresholds):
    plt.plot(thresholds,precisions[:-1],'b--',label='Precision')
    plt.plot(thresholds,recalls[:-1],'g-',label='Recall')
    plt.xlabel('Threshold')
    plt.legend(loc="center right", fontsize=16)
    plt.grid(True)                              # Not shown
    plt.axis([-50000, 50000, 0, 1]) 
    plt.figure(figsize=(8, 4)) 
```

```python
plot_precision_recall_vs_threshold(precisions,recalls,thresholds)
plt.show()
```

Vemos que la curva de precisi√≥n tiene m√°s accidentada (tiene mas picos o irrefularidades) que la curva de recall en el grafico. La raz√≥n es que la precisi√≥n a veces puede disminuir cuando aumenta el umbral (aunque en general aumentar√°). Para entender por qu√©, mire la Figura 3-3 y observe lo que sucede cuando comienza desde el umbral central y lo mueve solo un d√≠gito a la derecha: la precisi√≥n va de 4/5 (80%) a 3/4 (75%) ). Por otro lado, la recuperaci√≥n solo puede disminuir cuando se aumenta el umbral, lo que explica por qu√© su curva se ve suave.

Figura 3-3



Otra forma de selecionar una buena precision/recall trade-off es haciendo directamente un grafic de la precision contra el recall

```python
def plot_precision_vs_recall(precisions,recalls):
    plt.plot(recalls,precisions,'b-',linewidth=2)
    plt.xlabel("Recall", fontsize=16)
    plt.ylabel("Precision", fontsize=16)
    plt.axis([0, 1, 0, 1])
    plt.grid(True)                              # Not shown) 
```

```python
plot_precision_vs_recall(precisions,recalls)
plt.show()
```

Puede ver que la precisi√≥n realmente comienza a caer bruscamente alrededor del 80% de recall. Probablemente desee seleccionar una compensaci√≥n de precisi√≥n / recall justo antes de esa ca√≠da, por ejemplo, alrededor del 60% de recall. Pero, por supuesto, la elecci√≥n depende de su proyecto. Suponga que decide apuntar a una precisi√≥n del 90%. Busca el primer gr√°fico y encuentra que necesita usar un umbral de aproximadamente 8,000. Para ser m√°s precisos, puede buscar el umbral m√°s bajo que le brinde al menos un 90% de precisi√≥n (np.argmax () le dar√° el primer √≠ndice del valor m√°ximo, que en este caso significa el primer valor verdadero):

```python
threshold_90_precision = thresholds[np.argmax(precisions >= 0.90)]
```

Hagamos prediciones con el training set sin utilizar el metodo predic (por el momento) y usando el threshold que acabamos de encontrar, para eso usemos el siguiente codigo:

```python
y_train_pred_90 = (y_scores >= threshold_90_precision)
```

```python
precision_score(y_train_5,y_train_pred_90)
```

```python
recall_score(y_train_5,y_train_pred_90)
```

Como puede ver, es bastante f√°cil crear un clasificador con pr√°cticamente cualquier precisi√≥n que desee: simplemente establezca un umbral lo suficientemente alto y listo. Pero espera, no tan r√°pido. ¬°Un clasificador de alta precisi√≥n no es muy √∫til si su recuperaci√≥n es demasiado baja!

### The ROC Curve / Receiver Operating Characteristic, o Caracter√≠stica Operativa del Receptor

Para explicar qu√© es una curva ROC utilizaremos como ejemplo la concentraci√≥n de procalcitonina s√©rica (PCT), postulada como test diagn√≥stico para endocarditis infecciosa en sujetos con sospecha de presentar dicha patolog√≠a. La concentraci√≥n de PCT s√©rica corresponde a una escala continua, cuya mediana es significativamente diferente en los pacientes con y sin endocarditis infecciosa (6,56 vs 0. 44 ng/ml, respectivamente)5. Operativamente, para poder diagnosticar una endocarditis infecciosa a partir de un valor de PCT s√©rica es necesario establecer un punto de corte, consider√°ndose "positivos" aquellos valores mayores o iguales al punto de corte, y "negativos" aquellos valores menores al punto de corte. Al contrastar los resultados de la PCT s√©rica con los resultados de un est√°ndar de oro para el diagn√≥stico de endocarditis infecciosa (i.e. criterios de Duke) surgen cuatro posibles alternativas diagn√≥sticas:

1.	Pacientes que tienen una endocarditis infecciosa seg√∫n el est√°ndar de oro, y cuya medici√≥n de PCT s√©rica los diagnostica como "positivos". Estos pacientes reciben el nombre de "verdaderos positivos, VP".
2.	Pacientes que tienen una endocarditis infecciosa seg√∫n el est√°ndar de oro, sin embargo, su medici√≥n de PCT s√©rica los diagnostica como "negativos". Estos pacientes reciben el nombre de "falsos negativos, FN".
3. 	Pacientes que no tienen una endocarditis infecciosa seg√∫n el est√°ndar de oro, y cuya medici√≥n de PCT s√©rica los diagnostica como "negativos". Estos pacientes reciben el nombre de "verdaderos negativos, VN".
4.	Pacientes que no tienen una endocarditis infecciosa seg√∫n el est√°ndar de oro, sin embargo, su medici√≥n de PCT s√©rica los diagnostica como "positivos". Estos pacientes reciben el nombre de "falsos positivos, FP".


Conociendo el n√∫mero de sujetos de cada una de estas alternativas diagn√≥sticas es posible calcular la sensibilidad y especificidad para el punto de corte que les origina, seg√∫n las siguientes f√≥rmulas: sensibilidad = VP/(VP+FN) y especificidad = VN/(FP+VN) (Tabla 1). El lector notar√° que existen tantos puntos de corte posibles de PCT s√©rica como valores posee su escala de medici√≥n, y que cada punto de corte posee su respectiva sensibilidad y especificidad para el diagn√≥stico de endocarditis infecciosa. Un gr√°fico de curva ROC ilustra la sensibilidad y especificidad de cada uno de los posibles puntos de corte de un test diagn√≥stico cuya escala de medici√≥n es continua. La curva ROC se construye en base a la uni√≥n de distintos puntos de corte, correspondiendo el eje Y a la sensibilidad y el eje X a (1-especificidad) de cada uno de ellos. Ambos ejes incluyen valores entre 0 y 1 (0% a 100%). A modo de referencia, en todo gr√°fico de curva ROC se traza una l√≠nea desde el punto 0,0 al punto 1,1, llamada diagonal de referencia o l√≠nea de no-discriminaci√≥n (concepto a abordar m√°s adelante). La Figura 1 ilustra el gr√°fico de curva ROC de un test diagn√≥stico hipot√©tico.

Cerda, Jaime, & Cifuentes, Lorena. (2012). Uso de curvas ROC en investigaci√≥n cl√≠nica: Aspectos te√≥rico-pr√°cticos. Revista chilena de infectolog√≠a, 29(2), 138-141. https://dx.doi.org/10.4067/S0716-10182012000200003

La curva de receiver operating characteristic (ROC) es otra herramienta com√∫n utilizada con clasificadores binarios. Es muy similar a la curva de precisi√≥n / recall, pero en lugar de graficar la precisi√≥n frente al recall, la curva ROC grafica la tasa de verdaderos positivos (otro nombre para el recall) frente a la tasa de falsos positivos (FPR). El FPR es la proporci√≥n de instancias negativas que se clasifican incorrectamente como positivas. Es igual a la resta de 1 menos tasa negativa verdadera (TNR), que es la proporci√≥n de instancias negativas que se clasifican correctamente como negativas. El TNR tambi√©n se llama especificidad. Por lo tanto, la curva ROC traza la sensibilidad (recall) versus 1-especificidad. Para trazar la curva ROC, primero use la funci√≥n roc_curve () para calcular el TPR y FPR para varios valores de umbral:

G√©ron, Aur√©lien. Hands-On Machine Learning with Scikit-Learn, Keras, and TensorFlow (p. 97). O'Reilly Media. Edici√≥n de Kindle. 


```python
from sklearn.metrics import roc_curve
fpr, tpr, thresholds = roc_curve(y_train_5,y_scores) # fpr = false positive rate, tpr = true positive rate
```

```python
def plot_roc_curve(fpr,tpr,label = None):
    plt.plot(fpr,tpr,linewidth=2,label=label)
    plt.plot([0,1],[0,1],'k:')
    plt.grid(True) 
    plt.axis([0, 1, 0, 1]) 
    plt.xlabel('False positive rate (1 ‚Äì specificity)')
    plt.ylabel('True positive rate (Recall)')
```

```python
plot_roc_curve(fpr,tpr)
plt.show()
```

Una vez m√°s, hay un trade-off: cuanto mayor es el recall (TPR), m√°s falsos positivos (FPR) produce el clasificador. La l√≠nea de puntos (la diagonal) representa la curva ROC de un clasificador puramente aleatorio; un buen clasificador permanece lo m√°s lejos posible de esa l√≠nea (hacia la esquina superior izquierda).

Una forma de comparar clasificadores es midiendo el area bajo la curva(**AUC**). En una situcion perfecta donde nuestro clasificador distingue sin error a las dos clases el AUC es de 1 mientras que un clasificador malo casi aleatoreo tendra un AUC de 0,5 o menor.

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_train_5,y_scores)
```

## Multiclass Clasification

Mientras que los clasificadores binarios distinguen entre dos clases, los clasificadores multiclase (tambi√©n llamados clasificadores multinomiales) pueden distinguir entre m√°s de dos clases.

Algunos algoritmos (como Logistic Regression classifiers, Random Forest classifiers, and naive Bayes classifiers) son capaces de manejar varias clases de forma nativa. Otros (como los SGD Classifiers or Support Vector Machine classifiers) son clasificadores estrictamente binarios. Sin embargo, existen varias estrategias que puede utilizar para realizar una clasificaci√≥n multiclase con varios clasificadores binarios.


Una forma de crear un sistema que pueda clasificar las im√°genes de d√≠gitos en 10 clases (de 0 a 9) es entrenar 10 clasificadores binarios, uno para cada d√≠gito (un detector 0, un detector 1, un detector 2, etc. sobre). Luego, cuando desea clasificar una imagen, obtiene la puntuaci√≥n de decisi√≥n de cada clasificador para esa imagen y selecciona la clase cuyo clasificador genera la puntuaci√≥n m√°s alta. Esto se llama la estrategia one-versus-the-rest (OvR) strategy (also called one-versus-all).

Otra estrategia es entrenar un clasificador binario para cada par de d√≠gitos: uno para distinguir 0 y 1, otro para distinguir 0 y 2, otro para 1 y 2, y as√≠ sucesivamente. A esto se le llama la estrategia one-versus-one (OvO). Si hay N clases, necesita entrenar N √ó (N - 1) / 2 clasificadores. Para el problema de MNIST, ¬°esto significa entrenar 45 clasificadores binarios! Cuando quieras clasificar una imagen, tienes que pasar la imagen por los 45 clasificadores y ver qu√© clase gana m√°s duelos. La principal ventaja de OvO es que cada clasificador solo necesita ser entrenado con una parte del conjunto de entrenamiento para las dos clases que debe distinguir.

Algunos algoritmos (como los Support Vector Machine classifiers) escalan mal con el tama√±o del conjunto de entrenamiento. Para estos algoritmos, se prefiere OvO porque es m√°s r√°pido entrenar muchos clasificadores en conjuntos de entrenamiento peque√±os que entrenar pocos clasificadores en conjuntos de entrenamiento grandes. Sin embargo, para la mayor√≠a de los algoritmos de clasificaci√≥n binaria, se prefiere OvR.

Scikit-Learn detecta cuando intenta utilizar un algoritmo de clasificaci√≥n binaria para una tarea de clasificaci√≥n multiclase y ejecuta autom√°ticamente OvR u OvO, seg√∫n el algoritmo.

```python
from sklearn.svm import SVC
svm_clf = SVC()
svm_clf.fit(X_train,y_train)
svm_clf.predict([some_digit])
```

Vemos como el modelo SVC esta entrenado con el training set usando las clases objetivos que tiene todos los valores (y_train) del 0 al 9 (osea un clasificador multiclase) en vez de usar las clases objetico de solo 5/ 5 versus el resto (y_train_5).

Para este modelo la estrategia que usa por debajo es un OVO (One vs One) por lo que entren√≥ a 45 clasificadores binarios, obtuvo sus puntajes de decisi√≥n para la imagen y seleccion√≥ la clase que gan√≥ la mayor cantidad de duelos.

El an√°lisis de errores individuales tambi√©n puede ser una buena manera de obtener informaci√≥n sobre lo que est√° haciendo su clasificador y por qu√© est√° fallando, pero es m√°s dif√≠cil y requiere m√°s tiempo. Por ejemplo, tracemos ejemplos de 3 y 5 (la funci√≥n plot_digits () solo usa la funci√≥n imshow () de Matplotlib)

```python
# EXTRA
def plot_digits(instances, images_per_row=10, **options):
    size = 28
    images_per_row = min(len(instances), images_per_row)
    images = [instance.reshape(size,size) for instance in instances]
    n_rows = (len(instances) - 1) // images_per_row + 1
    row_images = []
    n_empty = n_rows * images_per_row - len(instances)
    images.append(np.zeros((size, size * n_empty)))
    for row in range(n_rows):
        rimages = images[row * images_per_row : (row + 1) * images_per_row]
        row_images.append(np.concatenate(rimages, axis=1))
    image = np.concatenate(row_images, axis=0)
    plt.imshow(image, cmap = mpl.cm.binary, **options)
    plt.axis("off")
```

```python
cl_a, cl_b = 3, 5
X_aa = X_train[(y_train == cl_a) & (y_train_pred == cl_a)]
X_ab = X_train[(y_train == cl_a) & (y_train_pred == cl_b)]
X_ba = X_train[(y_train == cl_b) & (y_train_pred == cl_a)]
X_bb = X_train[(y_train == cl_b) & (y_train_pred == cl_b)]

plt.figure(figsize=(8,8))
plt.subplot(221); plot_digits(X_aa[:25], images_per_row=5)
plt.subplot(222); plot_digits(X_ab[:25], images_per_row=5)
plt.subplot(223); plot_digits(X_ba[:25], images_per_row=5)
plt.subplot(224); plot_digits(X_bb[:25], images_per_row=5)
plt.show()
```

Los dos bloques de 5 √ó 5 de la izquierda muestran d√≠gitos clasificados como 3, y los dos bloques de 5 √ó 5 de la derecha muestran im√°genes clasificadas como 5. Algunos de los d√≠gitos en los que el clasificador se equivoca (es decir, en los bloques de abajo a la izquierda y de arriba a la derecha) est√°n tan mal escritos que incluso un humano tendr√≠a problemas para clasificarlos (por ejemplo, el 5 en la primera fila y la segunda columna realmente se ve como un mal escrito 3). Sin embargo, la mayor√≠a de las im√°genes clasificadas err√≥neamente nos parecen errores obvios y es dif√≠cil entender por qu√© el clasificador cometi√≥ los errores que cometi√≥. La raz√≥n es que usamos un clasificador SGDC simple, que es un modelo lineal. Todo lo que hace es asignar un peso por clase a cada p√≠xel, y cuando ve una nueva imagen, simplemente suma las intensidades de los p√≠xeles ponderados para obtener una puntuaci√≥n para cada clase. Entonces, dado que los 3 y los 5 difieren solo en unos pocos p√≠xeles, este modelo los confundir√° f√°cilmente. La principal diferencia entre 3 y 5 es la posici√≥n de la l√≠nea peque√±a que une la l√≠nea superior con el arco inferior. Si dibuja un 3 con la uni√≥n ligeramente desplazada hacia la izquierda, el clasificador podr√≠a clasificarlo como un 5 y viceversa. En otras palabras, este clasificador es bastante sensible al cambio y la rotaci√≥n de la imagen. Entonces, una forma de reducir la confusi√≥n 3/5 ser√≠a preprocesar las im√°genes para asegurarse de que est√©n bien centradas y no giradas demasiado. Esto probablemente tambi√©n ayudar√° a reducir otros errores.

## Multilabel Classification

Hasta ahora, cada instancia siempre se ha asignado a una sola clase. En algunos casos, es posible que desee que su clasificador genere varias clases para cada instancia. Considere un clasificador de reconocimiento facial: ¬øqu√© deber√≠a hacer si reconoce a varias personas en la misma imagen? Debe adjuntar una etiqueta por persona que reconozca. Digamos que el clasificador ha sido entrenado para reconocer tres caras, Alice, Bob y Charlie. Luego, cuando se muestra al clasificador una imagen de Alice y Charlie, deber√≠a generar [1, 0, 1] (que significa "Alice s√≠, Bob no, Charlie s√≠"). Un sistema de clasificaci√≥n de este tipo que genera m√∫ltiples etiquetas binarias se denomina **multilabel classification system**.

Hagamos un ejemplo simple:

```python
from sklearn.neighbors import KNeighborsClassifier
y_train_large = (y_train >= 7)
y_train_odd = (y_train % 2 == 1)
y_multilabel = np.c_[y_train_large,y_train_odd]

knn_clf = KNeighborsClassifier()
knn_clf.fit(X_train,y_multilabel)
```

```python
y_multilabel
```

Este c√≥digo crea una matriz y_multilabel que contiene dos etiquetas de destino para cada imagen de d√≠gito: la primera indica si el d√≠gito es grande o no (7, 8 o 9) y la segunda indica si es impar o no. Las siguientes l√≠neas crean una instancia de KNeighborsClassifier (que admite la clasificaci√≥n de m√∫ltiples etiquetas, aunque no todos los clasificadores lo hacen), y la entrenamos usando la matriz de m√∫ltiples objetivos. Ahora puede hacer una predicci√≥n y observar que genera dos etiquetas:

```python
knn_clf.predict([some_digit])
```

La prediccion fue correcta! The digit 5 is indeed not large (False) and odd (True).

Hay muchas formas de evaluar un clasificador de varias etiquetas, y la selecci√≥n de la m√©trica correcta realmente depende de su proyecto. Un enfoque es medir el puntaje F1 para cada etiqueta individual (o cualquier otra m√©trica de clasificador binario discutida anteriormente), luego simplemente calcular el puntaje promedio. Este c√≥digo calcula la puntuaci√≥n F1 promedio en todas las etiquetas:

## Multioutput Classification 

El √∫ltimo tipo de tarea de clasificaci√≥n que vamos a discutir aqu√≠ se llama  multioutput‚Äì multiclass classification (or simply multioutput classification). Es simplemente una generalizaci√≥n de la clasificaci√≥n de m√∫ltiples etiquetas donde cada etiqueta puede ser multiclase (es decir, puede tener m√°s de dos valores posibles). Para ilustrar esto, construyamos un sistema que elimine el ruido de las im√°genes. Tomar√° como entrada una imagen de d√≠gitos ruidosos y (con suerte) generar√° una imagen de d√≠gitos limpia, representada como una matriz de intensidades de p√≠xeles, al igual que las im√°genes MNIST. Observe que la salida del clasificador es de varias etiquetas (una etiqueta por p√≠xel) y cada etiqueta puede tener varios valores (la intensidad de los p√≠xeles var√≠a de 0 a 255). Por tanto, es un ejemplo de un sistema de clasificaci√≥n de salidas m√∫ltiples.

### Nota

La l√≠nea entre clasificaci√≥n y regresi√≥n a veces es borrosa, como en este ejemplo. Podr√≠a decirse que predecir la intensidad de los p√≠xeles es m√°s parecido a una regresi√≥n que a una clasificaci√≥n. Adem√°s, los sistemas de salida m√∫ltiple no se limitan a tareas de clasificaci√≥n; incluso podr√≠a tener un sistema que genere m√∫ltiples etiquetas por instancia, incluidas etiquetas de clase y etiquetas de valor.

Write a function that can shift an MNIST image in any direction (left, right, up, or down) by one pixel. 5 Then, for each image in the training set, create four shifted copies (one per direction) and add them to the training set. Finally, train your best model on this expanded training set and measure its accuracy on the test set. You should observe that your model performs even better now! This technique of artificially growing the training set is called data augmentation or training set expansion.
 

```python
from scipy.ndimage.interpolation import shift
import matplotlib.pyplot as plt

def shift_image(image,dx,dy):
    image = image.reshape((28,28))
    image_shifted = shift(image,[dy,dx],cval=0,mode='constant')
    return image_shifted.reshape([-1])

def show_image(image):
    return plt.imshow(image.reshape((28,28)), interpolation="nearest", cmap="Greys")
```

```python
image_down = shift_image(X_train[1],0,5)
image_up = shift_image(X_train[1],0,-5)
image_left = shift_image(X_train[1],-5,0)
image_right = shift_image(X_train[1],5,0)

plt.figure(figsize=(12,3))

plt.subplot(131)
plt.title("Original", fontsize=14)
show_image(X_train[1])
plt.subplot(132)
plt.title("Shifted down", fontsize=14)
show_image(image_down)
plt.subplot(133)
plt.title("Shifted up", fontsize=14)
show_image(image_up)
plt.show()
```

```python
X_train_augmented = [image for image in X_train]
y_train_augmented = [label for label in y_train]

for dx, dy in ((1, 0), (-1, 0), (0, 1), (0, -1)):
    for image, label in zip(X_train, y_train):
        X_train_augmented.append(shift_image(image, dx, dy))
        y_train_augmented.append(label)

X_train_augmented = np.array(X_train_augmented)
y_train_augmented = np.array(y_train_augmented)
```

```python
shuffle_idx = np.random.permutation(len(X_train_augmented))
X_train_augmented = X_train_augmented[shuffle_idx]
y_train_augmented = y_train_augmented[shuffle_idx]
```

```python
from sklearn.metrics import accuracy_score
y_pred_knn_2 = knn_clf_2.predict(X_test)
accuracy_score(y_test,y_pred_knn_2)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

num_pipeline = Pipeline([
    #('selec_numeric_params',DataFrameSlector(['Age','SibSp','Parch','Fare'])),
    ('imputer',SimpleImputer(strategy='median'))
])
```

```python
#Una forma de conseguir los valors categoricos mas frecuentes y con estos llenar los Nan
# Se usan series de pandas porque este es un parametro aceotado en la clase fillna de pandas
a = pd.Series([X_train[c].value_counts().index[0] for c in X_train[["Pclass", "Sex", "Embarked"]]],index=["Pclass", "Sex", "Embarked"])
a
```

```python
class MostFrequentImputer(BaseEstimator,TransformerMixin):
    def fit(self,X,y=None):
        self.most_frequent_ = pd.Series([X[c].value_counts().index[0] for c in X],index=X.columns)
        return self
    def transform(self,X,y=None):
        return X.fillna(self.most_frequent_)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
cat_pipeline = Pipeline([
    #('slect_categorical_params',DataFrameSlector(['Pclass','Sex','Embarked'])),
    ('imputer',MostFrequentImputer()),
    ('cat_encoder',OneHotEncoder(sparse=False))
])
```

```python
def pred_data(df):
    
    Num_params_titanic = df[['Age','SibSp','Parch','Fare']]
    Cat_params_titanic = df[['Pclass','Sex','Embarked']]
    
    X_with_out_age_na = num_pipeline.fit_transform(Num_params_titanic)
    df_train = pd.DataFrame(X_with_out_age_na,columns = num_params_titanic_list)
    
    df_train = pd.concat([df_train,Cat_params_titanic],axis=1) 
    df_train["AgeBucket"] = df_train["Age"] // 15 * 15
    df_train["RelativesOnboard"] = df_train["SibSp"] + df_train["Parch"]
    
    bins1 = [0,15,30,45,60,75]
    labels1 = ['Infant','Youths','Adult','Middle age','Elderly']
    df_train['AgeGroup'] = pd.cut(df_train['Age'], bins=bins1, labels=labels1, right=False,)
    
    bins2 = [0,1,2,4,10]
    labels = ['Alone','Accompanied','SmallFamily','BigFamily']
    df_train['RelativesOnBoardCategory'] = pd.cut(df_train['RelativesOnboard'], bins=bins2, labels=labels, right=False)
    df_train = df_train.drop(['SibSp','Parch','Age','RelativesOnboard','AgeBucket'],axis=1)
   
    num_params = list(df_train[['Fare']])
    cat_params = list(df_train[['Pclass','Sex','Embarked','AgeGroup','RelativesOnBoardCategory']])
    
    full_pipeline = ColumnTransformer([
    ('num_pipeline',num_pipeline,num_params),
    ('cat_pipeline',cat_pipeline,cat_params)
    ])
    
    X_test_pred = full_pipeline.fit_transform(df_train)
    
    X_test_sclaler = standard_scl.fit_transform(X_test_pred)
    return X_test_sclaler
```
